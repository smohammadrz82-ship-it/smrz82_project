```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data.csv')
df = pd.DataFrame(data)
print(df.columns)
```

# find average income or anyone in dataframe

```python
income_columns = [col for col in df.columns if 'income' in col]
df['Average Income'] = df[income_columns].mean(axis=1)
result = df[['Name', 'Average Income']]
print(income_columns)
plt.figure(figsize=(12, 6))
plt.bar(result['Name'], result['Average Income'], color='red')
plt.xlabel("names")
plt.ylabel("Average Income")
plt.title("Average Income for everyone")
plt.show()
```

```python
income_columns = [col for col in df.columns if 'income' in col]
months = [col.replace(' income', '') for col in income_columns]
for i in range(len(df)):
    plt.figure(figsize=(12, 6))
    plt.plot(months, df.loc[i, income_columns], label=df.loc[i, 'Name'])
    plt.xlabel("Months")
    plt.ylabel("Income")
    plt.title("incomefor anyone")
    plt.legend()
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```
