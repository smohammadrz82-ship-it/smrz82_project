```python
import pandas as pd
import numpy as np
import re
from scipy import stats
```

```python
data = pd.read_csv('data.csv')
df = pd.DataFrame(data)
print(df.columns)
```

# find mean and median of Age and Weight columns


```python
ages = df['Age'].to_list()
weights = df['Weight'].to_list()
weights = [int(re.findall(r'\d+', x)[0]) for x in weights]
print(ages)
print(weights)

mean_ages = np.mean(ages)
median_ages = np.median(ages)


mean_weights = np.mean(weights)
median_weights = np.median(weights)
print(f"mean of age : {mean_ages}")
print(f"median of age : {median_ages}")
print(f"mean of weight : {mean_weights}")
print(f"median of weights : {median_weights}")
```

# find mode of Height, Gender, Education

```python
heights = df['Height'].to_list()
genders = df["Gender"].to_list()
educations = df['Education'].to_list()
print(heights)
print(genders)
print(educations)

height_mode = df['Height'].mode()[0]
gender_mode = df['Gender'].mode()[0]
education_mode = df['Education'].mode()[0]
print(f"mode of height : {height_mode}")
print(f"mode of gender : {gender_mode}")
print(f"mode of education : {education_mode}")
```

# find range of Age, Height, Weight


```python
ages = df["Age"].to_list()
heights = df['Height'].to_list()
weights = df["Weight"].to_list()
print(ages)
print(heights)
print(weights)
weights = [int(re.findall(r'\d+', x)[0]) for x in weights]
print(weights)

age_range = np.ptp(ages)
height_range = np.ptp(heights)
weight_range = np.ptp(weights)
print(f"range of age : {age_range}")
print(f"range of height : {height_range}")
print(f"range of weight : {weight_range}")
```

# find variance and standard deviation of Age, Height, Weight

```python
ages = df["Age"].to_list()
heights = df['Height'].to_list()
weights = df["Weight"].to_list()
print(ages)
print(heights)
print(weights)
weights = [int(re.findall(r'\d+', x)[0]) for x in weights]
print(weights)

age_var = np.var(ages)
height_var = np.var(heights)
weight_var = np.var(weights)

age_std = np.std(ages)
height_std = np.std(heights)
weight_std = np.std(weights)

print(f"Variance of Age : {age_var}")
print(f"Variance of Height : {height_var}")
print(f"Variance of Weight : {weight_var}")

print(f"STD of Age : {age_std}")
print(f"STD of Height : {height_std}")
print(f"STD of Weight : {weight_std}")
```

# find Q1, Q2, Q3 of Age, Height, Weight

```python
ages = df["Age"].to_list()
heights = df['Height'].to_list()
weights = df["Weight"].to_list()
print(ages)
print(heights)
print(weights)
weights = [int(re.findall(r'\d+', x)[0]) for x in weights]
print(weights)

age_quartiles = np.percentile(ages, [25, 50, 75])
height_quartiles = np.percentile(heights, [25, 50, 75])
weight_quartiles = np.percentile(weights, [25, 50, 75])

print(f"Q1 of Age : {age_quartiles[0]}, Q2 of Age : {age_quartiles[1]}, Q3 of Age : {age_quartiles[2]}")
print(f"Q1 of Height: {height_quartiles[0]}, Q2 of Height : {height_quartiles[1]}, Q3 of Height : {height_quartiles[2]}")
print(f"Q1 of Weight : {weight_quartiles[0]}, Q2 of Weight : {weight_quartiles[1]}, Q3 of Weight : {weight_quartiles[2]}")
```

# find IQR of Age, Height, Weight

```python
ages = df["Age"].to_list()
heights = df['Height'].to_list()
weights = df["Weight"].to_list()
print(ages)
print(heights)
print(weights)
weights = [int(re.findall(r'\d+', x)[0]) for x in weights]
print(weights)

age_quartiles = np.percentile(ages, [25, 50, 75])
height_quartiles = np.percentile(heights, [25, 50, 75])
weight_quartiles = np.percentile(weights, [25, 50, 75])

age_iqr = age_quartiles[2] - age_quartiles[0]
height_iqr = height_quartiles[2] - height_quartiles[0]
weight_iqr = weight_quartiles[2] - weight_quartiles[0]

print(f"IQR of Age : {age_iqr}")
print(f"IQR of Height : {height_iqr}")
print(f"IQR of Weight : {weight_iqr}")
```
